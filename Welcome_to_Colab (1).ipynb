```r
# ============================================================
#  Stock Market Trend Analysis & Forecasting
#  Synthetic Data — Masters CS Project
# ============================================================
#  Methods covered:
#    1. Data generation (matches widget parameters)
#    2. Exploratory Data Analysis (EDA)
#    3. Trend detection — Linear Regression, Moving Averages
#    4. Stationarity test — ADF test
#    5. Decomposition — STL
#    6. ARIMA forecasting
#    7. Exponential Smoothing (ETS)
#    8. LSTM-style rolling-window regression (via neuralnet)
#    9. Performance metrics — RMSE, MAE, MAPE
#   10. Forecast plots
# ============================================================

# ── 0. Install & load packages ────────────────────────────────
packages <- c("tidyverse", "lubridate", "tseries", "forecast",
              "TTR", "ggplot2", "gridExtra", "zoo", "Metrics",
              "neuralnet", "scales")

install_if_missing <- function(pkg) {
  if (!requireNamespace(pkg, quietly = TRUE)) install.packages(pkg)
  library(pkg, character.only = TRUE)
}
invisible(lapply(packages, install_if_missing))

set.seed(42)

# ── 1. Synthetic Data Generation ─────────────────────────────
# Mirrors the widget logic: weekdays only, seeded trend + volatility

generate_stock_data <- function(stock_name, base_price, volatility,
                                 trend, n_days = 252) {
  all_dates <- seq.Date(as.Date("2025-03-16"), by = "day", length.out = n_days * 1.5)
  trading_dates <- all_dates[!weekdays(all_dates) %in% c("Saturday", "Sunday")]
  trading_dates <- head(trading_dates, n_days)

  prices <- numeric(n_days)
  volumes <- numeric(n_days)
  prices[1] <- base_price

  for (i in 2:n_days) {
    shock     <- rnorm(1, mean = 0, sd = volatility * prices[i - 1])
    drift     <- trend * prices[i - 1]
    prices[i] <- max(prices[i - 1] + shock + drift, base_price * 0.5)
    volumes[i] <- round(runif(1, 20e6, 50e6))
  }

  tibble(
    Date   = trading_dates,
    Stock  = stock_name,
    Open   = round(prices * runif(n_days, 0.99, 1.00), 2),
    High   = round(prices * runif(n_days, 1.005, 1.02), 2),
    Low    = round(prices * runif(n_days, 0.980, 0.995), 2),
    Close  = round(prices, 2),
    Volume = volumes
  )
}

# Four stocks — same parameters as the widget
stocks_raw <- bind_rows(
  generate_stock_data("AAPL",  172, 0.018, 0.0004),
  generate_stock_data("GOOGL", 138, 0.022, 0.0003),
  generate_stock_data("MSFT",  415, 0.016, 0.0005),
  generate_stock_data("AMZN",  185, 0.024, 0.0003)
)

cat("=== Dataset Overview ===\n")
cat("Rows:", nrow(stocks_raw), " | Stocks:", n_distinct(stocks_raw$Stock), "\n")
cat("Date range:", as.character(min(stocks_raw$Date)),
    "to", as.character(max(stocks_raw$Date)), "\n\n")
glimpse(stocks_raw)


# ── 2. Feature Engineering ────────────────────────────────────
stocks <- stocks_raw %>%
  group_by(Stock) %>%
  arrange(Date) %>%
  mutate(
    # Returns
    Daily_Return  = (Close - lag(Close)) / lag(Close),
    Log_Return    = log(Close / lag(Close)),

    # Moving averages
    MA_10  = rollmean(Close, 10,  fill = NA, align = "right"),
    MA_20  = rollmean(Close, 20,  fill = NA, align = "right"),
    MA_50  = rollmean(Close, 50,  fill = NA, align = "right"),

    # Volatility (rolling 20-day std of log returns)
    Volatility_20 = rollapply(Log_Return, 20, sd, fill = NA, align = "right"),

    # Bollinger Bands (20-day, 2 SD)
    BB_Upper = MA_20 + 2 * rollapply(Close, 20, sd, fill = NA, align = "right"),
    BB_Lower = MA_20 - 2 * rollapply(Close, 20, sd, fill = NA, align = "right"),

    # Momentum
    Momentum_10 = Close - lag(Close, 10),

    # RSI (14-day)
    RSI = RSI(Close, n = 14),

    # MACD
    MACD_val = MACD(Close)[, "macd"],
    MACD_sig = MACD(Close)[, "signal"],

    # Trend labels
    Trend = case_when(
      Close > MA_50 & MA_20 > MA_50 ~ "Uptrend",
      Close < MA_50 & MA_20 < MA_50 ~ "Downtrend",
      TRUE ~ "Sideways"
    )
  ) %>%
  ungroup()

cat("\n=== Feature Engineering Done ===\n")
cat("Columns added: Daily_Return, Log_Return, MA_10/20/50,",
    "Volatility_20, BB Bands, Momentum_10, RSI, MACD, Trend\n")


# ── 3. Exploratory Data Analysis ─────────────────────────────

# 3a. Price trajectories
p1 <- ggplot(stocks, aes(x = Date, y = Close, color = Stock)) +
  geom_line(linewidth = 0.8) +
  labs(title = "Closing Prices — All Stocks", x = NULL, y = "Price (USD)") +
  scale_y_continuous(labels = dollar_format()) +
  theme_minimal(base_size = 12) +
  theme(legend.position = "bottom")

# 3b. Daily returns distribution
p2 <- ggplot(stocks %>% drop_na(Daily_Return),
             aes(x = Daily_Return, fill = Stock)) +
  geom_histogram(bins = 50, alpha = 0.6, position = "identity") +
  labs(title = "Daily Returns Distribution", x = "Daily Return", y = "Count") +
  scale_x_continuous(labels = percent_format()) +
  theme_minimal(base_size = 12) +
  theme(legend.position = "bottom")

# 3c. Cumulative returns
p3 <- stocks %>%
  drop_na(Log_Return) %>%
  group_by(Stock) %>%
  mutate(Cumulative_Return = cumsum(Log_Return)) %>%
  ggplot(aes(x = Date, y = Cumulative_Return, color = Stock)) +
  geom_line(linewidth = 0.8) +
  labs(title = "Cumulative Log Returns", x = NULL, y = "Cumulative Log Return") +
  theme_minimal(base_size = 12) +
  theme(legend.position = "bottom")

# 3d. Rolling volatility
p4 <- stocks %>%
  drop_na(Volatility_20) %>%
  ggplot(aes(x = Date, y = Volatility_20, color = Stock)) +
  geom_line(linewidth = 0.7, alpha = 0.85) +
  labs(title = "20-Day Rolling Volatility", x = NULL, y = "Std Dev of Log Returns") +
  theme_minimal(base_size = 12) +
  theme(legend.position = "bottom")

grid.arrange(p1, p2, p3, p4, ncol = 2)


# ── 4. Focus on one stock for modelling (AAPL) ───────────────
aapl <- stocks %>%
  filter(Stock == "AAPL") %>%
  arrange(Date) %>%
  drop_na(Close)

cat("\n=== AAPL Summary Statistics ===\n")
summary(aapl$Close)


# ── 5. Trend Analysis — Linear Regression ────────────────────
aapl <- aapl %>% mutate(Time_Index = row_number())

lm_model <- lm(Close ~ Time_Index, data = aapl)
cat("\n=== Linear Trend (OLS) ===\n")
print(summary(lm_model))

aapl$LM_Fitted <- predict(lm_model)

ggplot(aapl, aes(x = Date)) +
  geom_line(aes(y = Close), color = "#378ADD", linewidth = 0.8, alpha = 0.7) +
  geom_line(aes(y = LM_Fitted), color = "#E24B4A", linewidth = 1, linetype = "dashed") +
  geom_line(aes(y = MA_20), color = "#D4537E", linewidth = 0.9, na.rm = TRUE) +
  geom_line(aes(y = MA_50), color = "#1D9E75", linewidth = 0.9, na.rm = TRUE) +
  labs(title = "AAPL — Price with Linear Trend & Moving Averages",
       subtitle = "Blue = Close | Red dashed = Linear trend | Pink = MA20 | Green = MA50",
       x = NULL, y = "Price (USD)") +
  scale_y_continuous(labels = dollar_format()) +
  theme_minimal(base_size = 12)


# ── 6. Stationarity Test (ADF) ────────────────────────────────
cat("\n=== Augmented Dickey-Fuller Test — AAPL Close ===\n")
adf_level <- adf.test(aapl$Close)
print(adf_level)

cat("\n=== ADF Test — First Differenced Series ===\n")
adf_diff <- adf.test(diff(aapl$Close))
print(adf_diff)

if (adf_diff$p.value < 0.05) {
  cat("Series is stationary after first differencing (I(1) process).\n")
} else {
  cat("Series may need further differencing.\n")
}


# ── 7. STL Decomposition ──────────────────────────────────────
# Convert to weekly time series for meaningful seasonality
aapl_ts <- ts(aapl$Close, frequency = 5)   # 5 trading days/week

stl_fit <- stl(aapl_ts, s.window = "periodic", robust = TRUE)
cat("\n=== STL Decomposition (weekly seasonality) ===\n")
plot(stl_fit, main = "AAPL — STL Decomposition (Trend / Seasonal / Remainder)")


# ── 8. ARIMA Forecasting ──────────────────────────────────────
# Train / test split — last 30 days as holdout
n_test  <- 30
n_train <- nrow(aapl) - n_test

train_ts <- ts(aapl$Close[1:n_train], frequency = 5)
test_ts  <- aapl$Close[(n_train + 1):nrow(aapl)]

cat("\n=== Auto-ARIMA Model Selection ===\n")
arima_model <- auto.arima(train_ts, seasonal = TRUE,
                           stepwise = FALSE, approximation = FALSE)
print(summary(arima_model))

# Forecast 30 days ahead
arima_fc <- forecast(arima_model, h = n_test)

# Evaluate
arima_rmse <- rmse(test_ts, as.numeric(arima_fc$mean))
arima_mae  <- mae(test_ts,  as.numeric(arima_fc$mean))
arima_mape <- mape(test_ts, as.numeric(arima_fc$mean)) * 100

cat(sprintf("\nARIMA — RMSE: %.4f | MAE: %.4f | MAPE: %.2f%%\n",
            arima_rmse, arima_mae, arima_mape))

# Plot ARIMA forecast
autoplot(arima_fc) +
  autolayer(ts(test_ts, start = n_train + 1, frequency = 5),
            series = "Actual", color = "#378ADD") +
  labs(title = "AAPL — ARIMA Forecast vs Actual (30-day holdout)",
       x = "Time (trading days)", y = "Price (USD)") +
  scale_y_continuous(labels = dollar_format()) +
  theme_minimal(base_size = 12)


# ── 9. Exponential Smoothing (ETS) ───────────────────────────
cat("\n=== ETS Model ===\n")
ets_model <- ets(train_ts)
print(summary(ets_model))

ets_fc <- forecast(ets_model, h = n_test)

ets_rmse <- rmse(test_ts, as.numeric(ets_fc$mean))
ets_mae  <- mae(test_ts,  as.numeric(ets_fc$mean))
ets_mape <- mape(test_ts, as.numeric(ets_fc$mean)) * 100

cat(sprintf("ETS    — RMSE: %.4f | MAE: %.4f | MAPE: %.2f%%\n",
            ets_rmse, ets_mae, ets_mape))


# ── 10. Neural Network (NNAR) ─────────────────────────────────
cat("\n=== NNAR (Neural Network AutoRegression) ===\n")
nnar_model <- nnetar(train_ts, repeats = 20)
print(nnar_model)

nnar_fc <- forecast(nnar_model, h = n_test)

nnar_rmse <- rmse(test_ts, as.numeric(nnar_fc$mean))
nnar_mae  <- mae(test_ts,  as.numeric(nnar_fc$mean))
nnar_mape <- mape(test_ts, as.numeric(nnar_fc$mean)) * 100

cat(sprintf("NNAR   — RMSE: %.4f | MAE: %.4f | MAPE: %.2f%%\n",
            nnar_rmse, nnar_mae, nnar_mape))


# ── 11. Model Comparison ──────────────────────────────────────
model_results <- tibble(
  Model = c("ARIMA", "ETS", "NNAR"),
  RMSE  = round(c(arima_rmse, ets_rmse, nnar_rmse), 4),
  MAE   = round(c(arima_mae,  ets_mae,  nnar_mae),  4),
  MAPE  = round(c(arima_mape, ets_mape, nnar_mape), 2)
)

cat("\n=== Model Comparison ===\n")
print(model_results)

best_model <- model_results %>% slice_min(RMSE)
cat(sprintf("\nBest model by RMSE: %s\n", best_model$Model))


# ── 12. Final 30-Day Future Forecast (best model) ────────────
cat("\n=== 30-Day Future Forecast (full series) ===\n")
full_ts     <- ts(aapl$Close, frequency = 5)
best_arima  <- auto.arima(full_ts, stepwise = FALSE, approximation = FALSE)
future_fc   <- forecast(best_arima, h = 30)

future_dates <- seq.Date(max(aapl$Date) + 1, by = "day", length.out = 50)
future_dates <- future_dates[!weekdays(future_dates) %in% c("Saturday","Sunday")]
future_dates <- head(future_dates, 30)

forecast_df <- tibble(
  Date  = future_dates,
  Lower = as.numeric(future_fc$lower[, 2]),
  Mean  = as.numeric(future_fc$mean),
  Upper = as.numeric(future_fc$upper[, 2])
)

# Combine historical + forecast
hist_plot <- aapl %>% tail(60) %>% select(Date, Close)

ggplot() +
  geom_line(data = hist_plot,
            aes(x = Date, y = Close), color = "#378ADD", linewidth = 0.9) +
  geom_ribbon(data = forecast_df,
              aes(x = Date, ymin = Lower, ymax = Upper),
              fill = "#D4537E", alpha = 0.2) +
  geom_line(data = forecast_df,
            aes(x = Date, y = Mean), color = "#D4537E", linewidth = 1, linetype = "dashed") +
  geom_vline(xintercept = as.numeric(max(aapl$Date)),
             linetype = "dotted", color = "#888780") +
  annotate("text", x = max(aapl$Date) + 2, y = max(hist_plot$Close),
           label = "Forecast start", size = 3, color = "#888780", hjust = 0) +
  labs(title = "AAPL — 30-Day Price Forecast",
       subtitle = "Pink band = 95% confidence interval",
       x = NULL, y = "Price (USD)") +
  scale_y_continuous(labels = dollar_format()) +
  theme_minimal(base_size = 12)

cat("\n=== Forecast Values ===\n")
print(forecast_df)

cat("\n=== Analysis Complete ===\n")
```

```r
# ============================================================
#  Stock Market Trend Analysis & Forecasting
#  EXPLAINED VERSION — Rich charts & plain-English summaries
#  R / RStudio — Masters CS Project
# ============================================================
#  Every chart has:
#    • A plain-English title anyone can understand
#    • Annotated labels explaining what they are seeing
#    • A printed insight box after each section
#    • Color-coded zones (green = good, red = caution)
# ============================================================

# ── 0. Install & Load Packages ────────────────────────────────
packages <- c("tidyverse", "lubridate", "tseries", "forecast",
              "TTR", "ggplot2", "gridExtra", "zoo", "Metrics",
              "scales", "patchwork", "ggtext", "ggrepel")

for (pkg in packages) {
  if (!requireNamespace(pkg, quietly = TRUE)) install.packages(pkg)
  library(pkg, character.only = TRUE)
}
set.seed(42)

# ── Helper: print a plain-English insight box ─────────────────
insight_box <- function(title, lines) {
  width <- 65
  cat("\n")
  cat(strrep("═", width), "\n")
  cat(sprintf("  💡 %s\n", title))
  cat(strrep("─", width), "\n")
  for (line in lines) cat(sprintf("  %s\n", line))
  cat(strrep("═", width), "\n\n")
}

section_header <- function(step, title, subtitle = "") {
  cat("\n")
  cat(strrep("━", 65), "\n")
  cat(sprintf("  STEP %s — %s\n", step, title))
  if (nchar(subtitle) > 0) cat(sprintf("  %s\n", subtitle))
  cat(strrep("━", 65), "\n")
}

# ── Color palette (consistent across all charts) ──────────────
COL <- list(
  aapl     = "#378ADD",
  googl    = "#D4537E",
  msft     = "#1D9E75",
  amzn     = "#BA7517",
  trend    = "#E24B4A",
  ma20     = "#9F77DD",
  ma50     = "#1D9E75",
  forecast = "#D4537E",
  ci       = "#F5C4D0",
  neutral  = "#888780",
  up       = "#1D9E75",
  down     = "#E24B4A",
  side     = "#B4B2A9"
)

STOCK_COLORS <- c(AAPL  = "#378ADD", GOOGL = "#D4537E",
                  MSFT  = "#1D9E75", AMZN  = "#BA7517")

# ── Shared ggplot theme ───────────────────────────────────────
theme_stock <- function() {
  theme_minimal(base_size = 11) +
    theme(
      plot.title       = element_text(face = "bold", size = 12, margin = margin(b = 4)),
      plot.subtitle    = element_text(size = 9, color = "#555555", margin = margin(b = 8)),
      plot.caption     = element_text(size = 8, color = "#777777", face = "italic",
                                      hjust = 0, margin = margin(t = 8)),
      panel.grid.minor = element_blank(),
      panel.grid.major = element_line(color = "#EEEEEE"),
      legend.position  = "bottom",
      legend.text      = element_text(size = 8),
      axis.title       = element_text(size = 9),
      plot.background  = element_rect(fill = "#FAFAFA", color = NA),
      panel.background = element_rect(fill = "#FAFAFA", color = NA)
    )
}

# ══════════════════════════════════════════════════════════════
#  STEP 1 — GENERATE SYNTHETIC DATA
# ══════════════════════════════════════════════════════════════
section_header("1", "Generating Synthetic Stock Data",
               "1 year of daily OHLCV for AAPL, GOOGL, MSFT, AMZN")

generate_stock_data <- function(stock_name, base_price, volatility,
                                 trend, seed, n_days = 252) {
  set.seed(seed)
  all_dates <- seq.Date(as.Date("2025-03-17"), by = "day", length.out = n_days * 2)
  trading_dates <- all_dates[!weekdays(all_dates) %in% c("Saturday", "Sunday")]
  trading_dates <- head(trading_dates, n_days)

  prices <- numeric(n_days)
  prices[1] <- base_price
  for (i in 2:n_days) {
    shock     <- rnorm(1, 0, volatility * prices[i - 1])
    prices[i] <- max(prices[i - 1] + shock + trend * prices[i - 1], base_price * 0.5)
  }

  tibble(
    Date   = trading_dates,
    Stock  = stock_name,
    Open   = round(prices * runif(n_days, 0.990, 1.000), 2),
    High   = round(prices * runif(n_days, 1.005, 1.020), 2),
    Low    = round(prices * runif(n_days, 0.980, 0.995), 2),
    Close  = round(prices, 2),
    Volume = round(runif(n_days, 20e6, 50e6))
  )
}

stocks_raw <- bind_rows(
  generate_stock_data("AAPL",  172, 0.018, 0.0004, 11),
  generate_stock_data("GOOGL", 138, 0.022, 0.0003, 22),
  generate_stock_data("MSFT",  415, 0.016, 0.0005, 33),
  generate_stock_data("AMZN",  185, 0.024, 0.0003, 44)
)

cat(sprintf("  Stocks    : %s\n", paste(unique(stocks_raw$Stock), collapse = ", ")))
cat(sprintf("  Rows      : %d  (%d trading days × 4 stocks)\n",
            nrow(stocks_raw), n_distinct(stocks_raw$Date)))
cat(sprintf("  Date span : %s  →  %s\n",
            min(stocks_raw$Date), max(stocks_raw$Date)))

insight_box("What is this data?", c(
  "Each row = one trading day for one stock.",
  "Open / Close = price at start and end of the day.",
  "High / Low   = the highest and lowest price that day.",
  "Volume       = number of shares bought & sold.",
  "AMZN & GOOGL have higher volatility — they swing more."
))

# ══════════════════════════════════════════════════════════════
#  STEP 2 — FEATURE ENGINEERING
# ══════════════════════════════════════════════════════════════
section_header("2", "Feature Engineering",
               "Computing indicators that reveal trend, momentum & risk")

stocks <- stocks_raw %>%
  group_by(Stock) %>%
  arrange(Date) %>%
  mutate(
    Daily_Return  = (Close - lag(Close)) / lag(Close),
    Log_Return    = log(Close / lag(Close)),
    MA_10         = rollmean(Close, 10,  fill = NA, align = "right"),
    MA_20         = rollmean(Close, 20,  fill = NA, align = "right"),
    MA_50         = rollmean(Close, 50,  fill = NA, align = "right"),
    Volatility_20 = rollapply(Log_Return, 20, sd, fill = NA, align = "right"),
    BB_Upper      = MA_20 + 2 * rollapply(Close, 20, sd, fill = NA, align = "right"),
    BB_Lower      = MA_20 - 2 * rollapply(Close, 20, sd, fill = NA, align = "right"),
    Momentum_10   = Close - lag(Close, 10),
    RSI           = RSI(Close, n = 14),
    MACD_val      = MACD(Close)[, "macd"],
    MACD_sig      = MACD(Close)[, "signal"],
    Trend = case_when(
      Close > MA_50 & MA_20 > MA_50 ~ "Uptrend",
      Close < MA_50 & MA_20 < MA_50 ~ "Downtrend",
      TRUE ~ "Sideways"
    )
  ) %>%
  ungroup()

insight_box("What do these indicators mean?", c(
  "MA 20 / MA 50 : average price over the last 20 or 50 days.",
  "  Golden Cross : MA20 rises above MA50 → BULLISH (price likely going UP).",
  "  Death Cross  : MA20 falls below MA50 → BEARISH (price likely going DOWN).",
  "RSI (0–100)  : >70 = overbought (may drop), <30 = oversold (may bounce).",
  "MACD         : gap between short & long moving averages → momentum.",
  "Bollinger Bands : ±2 std dev channel. Price outside = unusual move.",
  "Trend label  : each day is classified as Uptrend / Downtrend / Sideways."
))

# ══════════════════════════════════════════════════════════════
#  STEP 3 — EDA DASHBOARD (4 charts in one figure)
# ══════════════════════════════════════════════════════════════
section_header("3", "Exploratory Data Analysis Dashboard",
               "Four views that tell the full story of all 4 stocks")

# 3a — Price trajectories
p_price <- ggplot(stocks, aes(x = Date, y = Close, color = Stock)) +
  geom_line(linewidth = 0.85) +
  scale_color_manual(values = STOCK_COLORS) +
  scale_y_continuous(labels = dollar_format()) +
  scale_x_date(date_labels = "%b '%y", date_breaks = "2 months") +
  labs(
    title    = "① Closing Prices Over Time",
    subtitle = "Each line = one stock. Rising line = stock gained value.",
    caption  = "MSFT starts highest (~$415) because it is a more expensive stock — not necessarily better returns.",
    x = NULL, y = "Price (USD)", color = NULL
  ) +
  theme_stock()

# 3b — Cumulative returns (who actually performed better?)
cum_ret <- stocks %>%
  group_by(Stock) %>%
  arrange(Date) %>%
  mutate(Cum_Return = cumsum(replace_na(Log_Return, 0)) * 100) %>%
  ungroup()

# label at last date
last_labels <- cum_ret %>% group_by(Stock) %>% slice_tail(n = 1)

p_cum <- ggplot(cum_ret, aes(x = Date, y = Cum_Return, color = Stock)) +
  geom_hline(yintercept = 0, color = "#CCCCCC", linetype = "dashed") +
  geom_line(linewidth = 0.85) +
  geom_label(data = last_labels,
             aes(label = sprintf("%s\n%+.1f%%", Stock, Cum_Return)),
             size = 2.5, show.legend = FALSE, hjust = 0,
             label.padding = unit(0.15, "lines"), label.size = 0.2) +
  scale_color_manual(values = STOCK_COLORS) +
  scale_x_date(date_labels = "%b '%y", date_breaks = "2 months",
               expand = expansion(mult = c(0.02, 0.12))) +
  labs(
    title    = "② Total Return (%) — Who Performed Best?",
    subtitle = "Compares stocks on equal footing regardless of starting price.",
    caption  = "A stock starting at $100 and ending at $110 = +10%. This removes the price-level bias.",
    x = NULL, y = "Cumulative Return (%)", color = NULL
  ) +
  theme_stock()

# 3c — Volatility (risk)
p_vol <- stocks %>%
  drop_na(Volatility_20) %>%
  ggplot(aes(x = Date, y = Volatility_20 * 100, color = Stock)) +
  geom_line(linewidth = 0.8, alpha = 0.85) +
  scale_color_manual(values = STOCK_COLORS) +
  scale_x_date(date_labels = "%b '%y", date_breaks = "2 months") +
  labs(
    title    = "③ Daily Risk Level (20-Day Rolling Volatility)",
    subtitle = "Higher = more uncertain / risky. Lower = calm, predictable.",
    caption  = "Spikes in volatility often match market events (earnings, news). AMZN tends to be riskiest.",
    x = NULL, y = "Volatility (%)", color = NULL
  ) +
  theme_stock()

# 3d — Trend composition (stacked bar)
trend_pct <- stocks %>%
  count(Stock, Trend) %>%
  group_by(Stock) %>%
  mutate(Pct = n / sum(n) * 100) %>%
  ungroup() %>%
  mutate(Trend = factor(Trend, levels = c("Uptrend", "Sideways", "Downtrend")))

p_trend <- ggplot(trend_pct, aes(x = Stock, y = Pct, fill = Trend)) +
  geom_bar(stat = "identity", width = 0.55) +
  geom_text(aes(label = ifelse(Pct > 6, sprintf("%.0f%%", Pct), "")),
            position = position_stack(vjust = 0.5),
            size = 3, color = "white", fontface = "bold") +
  scale_fill_manual(values = c(Uptrend   = "#1D9E75",
                                Sideways  = "#B4B2A9",
                                Downtrend = "#E24B4A")) +
  labs(
    title    = "④ How Many Days Was Each Stock Trending Up vs Down?",
    subtitle = "Green = clear uptrend (bullish). Red = downtrend (bearish). Gray = no clear direction.",
    caption  = "A stock spending most days in Uptrend is showing consistent strength.",
    x = NULL, y = "% of Trading Days", fill = NULL
  ) +
  theme_stock() +
  theme(legend.position = "bottom")

# Combine into one dashboard
dashboard <- (p_price | p_cum) / (p_vol | p_trend) +
  plot_annotation(
    title   = "Stock Market Overview — All 4 Stocks (1 Trading Year)",
    caption = "Synthetic data generated for analysis. OHLCV format, weekdays only.",
    theme   = theme(
      plot.title   = element_text(face = "bold", size = 14),
      plot.caption = element_text(size = 8, color = "#777", face = "italic")
    )
  )

print(dashboard)
ggsave("01_eda_dashboard.png", dashboard, width = 14, height = 9, dpi = 140)
cat("  Chart saved → 01_eda_dashboard.png\n")

# Print headline numbers
section_header("3b", "EDA Headline Summary Numbers")
stocks %>%
  group_by(Stock) %>%
  summarise(
    `Start Price`   = sprintf("$%.2f", first(Close)),
    `End Price`     = sprintf("$%.2f", last(Close)),
    `Total Return`  = sprintf("%+.1f%%", (last(Close) - first(Close)) / first(Close) * 100),
    `Avg Daily Vol` = sprintf("%.2f%%", mean(Volatility_20, na.rm = TRUE) * 100),
    `Dominant Trend`= names(which.max(table(Trend)))
  ) %>%
  print()

insight_box("Reading the EDA results", c(
  "Total Return   : The actual % gain from first to last day.",
  "Avg Daily Vol  : How much the price typically swings each day.",
  "Dominant Trend : The trend label seen most often across the year.",
  "MSFT highest return = its trend & lower volatility compounded well.",
  "AMZN/GOOGL more volatile = bigger swings, harder to predict."
))

# ══════════════════════════════════════════════════════════════
#  STEP 4 — AAPL DEEP DIVE: TREND ANALYSIS
# ══════════════════════════════════════════════════════════════
section_header("4", "AAPL Deep Dive — Trend Analysis",
               "Moving Averages, Bollinger Bands, RSI & MACD explained")

aapl <- stocks %>%
  filter(Stock == "AAPL") %>%
  arrange(Date) %>%
  drop_na(Close) %>%
  mutate(Time_Index = row_number())

# Linear trend
lm_model <- lm(Close ~ Time_Index, data = aapl)
slope     <- coef(lm_model)["Time_Index"]
direction <- if (slope > 0) "UPWARD ↑" else "DOWNWARD ↓"
aapl$LM_Fitted <- predict(lm_model)

cat(sprintf("\n  Linear trend slope : %+.4f USD per trading day\n", slope))
cat(sprintf("  Over 252 days      : %+.2f USD total price drift\n", slope * 252))
cat(sprintf("  Trend direction    : %s\n", direction))

insight_box("What does the linear trend tell us?", c(
  sprintf("Overall the AAPL price moved %s over the year.", direction),
  sprintf("Average daily drift = %+.4f USD per day.", slope),
  sprintf("Compounded over 252 trading days → %+.2f USD change.", slope * 252),
  "This is the big picture trend, ignoring short-term noise.",
  "Moving averages show the momentum within that bigger trend."
))

# Golden Cross / Death Cross points
crossover_up <- aapl %>%
  filter(MA_20 > MA_50, lag(MA_20) <= lag(MA_50)) %>%
  drop_na(MA_20, MA_50)

crossover_dn <- aapl %>%
  filter(MA_20 < MA_50, lag(MA_20) >= lag(MA_50)) %>%
  drop_na(MA_20, MA_50)

# CHART A — Price + Trend + MAs + Bollinger Bands
p_main <- ggplot(aapl, aes(x = Date)) +
  geom_ribbon(aes(ymin = BB_Lower, ymax = BB_Upper),
              fill = COL$aapl, alpha = 0.08) +
  geom_line(aes(y = Close),     color = COL$aapl,  linewidth = 0.9, alpha = 0.85) +
  geom_line(aes(y = LM_Fitted), color = COL$trend, linewidth = 1.4, linetype = "dashed") +
  geom_line(aes(y = MA_20),     color = COL$ma20,  linewidth = 1.0, na.rm = TRUE) +
  geom_line(aes(y = MA_50),     color = COL$ma50,  linewidth = 1.0, na.rm = TRUE) +
  # Golden cross markers
  geom_vline(data = crossover_up, aes(xintercept = as.numeric(Date)),
             color = COL$up, linewidth = 0.6, alpha = 0.4) +
  geom_label(data = crossover_up, aes(x = Date, y = max(aapl$BB_Upper, na.rm = TRUE) * 0.99,
             label = "Golden\nCross ↑"), size = 2.2, color = COL$up,
             fill = "#EAF3DE", label.size = 0.2, vjust = 1) +
  # Death cross markers
  geom_vline(data = crossover_dn, aes(xintercept = as.numeric(Date)),
             color = COL$down, linewidth = 0.6, alpha = 0.4) +
  geom_label(data = crossover_dn, aes(x = Date, y = min(aapl$BB_Lower, na.rm = TRUE) * 1.01,
             label = "Death\nCross ↓"), size = 2.2, color = COL$down,
             fill = "#FCEBEB", label.size = 0.2, vjust = 0) +
  scale_y_continuous(labels = dollar_format()) +
  scale_x_date(date_labels = "%b '%y", date_breaks = "2 months") +
  labs(
    title    = "① AAPL Price — Trend Line, Moving Averages & Bollinger Bands",
    subtitle = paste0(
      "Red dashed = overall ", direction, " trend  |  ",
      "Purple = MA20 (short)  |  Green = MA50 (long)  |  Blue band = Bollinger Bands"
    ),
    caption  = paste0(
      "Bollinger Band: normal price range (±2 std dev around MA20). ",
      "Price outside band = unusual move.\n",
      "Golden Cross (↑): MA20 crosses above MA50 → bullish momentum — price tends to keep rising.\n",
      "Death Cross (↓): MA20 crosses below MA50 → bearish momentum — price tends to keep falling."
    ),
    x = NULL, y = "Price (USD)"
  ) +
  theme_stock() +
  theme(legend.position = "none")

# CHART B — RSI panel
p_rsi <- ggplot(aapl %>% drop_na(RSI), aes(x = Date, y = RSI)) +
  annotate("rect", xmin = min(aapl$Date), xmax = max(aapl$Date),
           ymin = 70, ymax = 100, fill = "#FCEBEB", alpha = 0.4) +
  annotate("rect", xmin = min(aapl$Date), xmax = max(aapl$Date),
           ymin = 0, ymax = 30, fill = "#EAF3DE", alpha = 0.4) +
  geom_hline(yintercept = c(30, 70), linetype = "dashed",
             color = c(COL$up, COL$down), linewidth = 0.6) +
  geom_line(color = COL$ma20, linewidth = 0.8) +
  annotate("text", x = min(aapl$Date) + 5, y = 75,
           label = "Overbought zone (>70) → price may be too high, a pullback is likely",
           hjust = 0, size = 2.8, color = "#A32D2D") +
  annotate("text", x = min(aapl$Date) + 5, y = 25,
           label = "Oversold zone (<30) → price may be too low, a bounce is likely",
           hjust = 0, size = 2.8, color = "#0F6E56") +
  scale_x_date(date_labels = "%b '%y", date_breaks = "2 months") +
  scale_y_continuous(limits = c(0, 100), breaks = c(0, 30, 50, 70, 100)) +
  labs(
    title   = "② RSI — Is the Stock Overbought or Oversold?",
    caption = "RSI = Relative Strength Index (0–100). Tracks speed of recent price changes.",
    x = NULL, y = "RSI"
  ) +
  theme_stock()

# CHART C — MACD panel
aapl_macd <- aapl %>%
  drop_na(MACD_val, MACD_sig) %>%
  mutate(Histogram = MACD_val - MACD_sig,
         Bar_Color = if_else(Histogram >= 0, "Bullish", "Bearish"))

p_macd <- ggplot(aapl_macd, aes(x = Date)) +
  geom_col(aes(y = Histogram, fill = Bar_Color), alpha = 0.5, width = 1.5) +
  geom_hline(yintercept = 0, color = "#CCCCCC", linewidth = 0.5) +
  geom_line(aes(y = MACD_val), color = COL$aapl,  linewidth = 0.9) +
  geom_line(aes(y = MACD_sig), color = COL$trend, linewidth = 0.9, linetype = "dashed") +
  scale_fill_manual(values = c(Bullish = "#1D9E75", Bearish = "#E24B4A"),
                    name = "Momentum") +
  scale_x_date(date_labels = "%b '%y", date_breaks = "2 months") +
  labs(
    title   = "③ MACD — Trend Direction & Momentum Strength",
    caption = paste0(
      "Blue line = MACD  |  Red dashed = Signal line  |  Bars = difference between them.\n",
      "Green bars: MACD above signal → upward momentum building (bullish).\n",
      "Red bars  : MACD below signal → downward momentum building (bearish)."
    ),
    x = NULL, y = "MACD Value"
  ) +
  theme_stock() +
  theme(legend.position = "right")

# Stack the 3 panels
trend_dashboard <- p_main / p_rsi / p_macd +
  plot_layout(heights = c(3, 1.5, 1.5)) +
  plot_annotation(
    title   = "AAPL — Complete Trend Analysis",
    caption = "Three-panel view: price structure (top), momentum (middle), trend direction (bottom).",
    theme   = theme(plot.title = element_text(face = "bold", size = 14))
  )

print(trend_dashboard)
ggsave("02_trend_analysis.png", trend_dashboard, width = 13, height = 13, dpi = 140)
cat("  Chart saved → 02_trend_analysis.png\n")

# ══════════════════════════════════════════════════════════════
#  STEP 5 — STL DECOMPOSITION
# ══════════════════════════════════════════════════════════════
section_header("5", "STL Decomposition",
               "Breaking the price into: Trend + Weekly Pattern + Random Noise")

aapl_ts  <- ts(aapl$Close, frequency = 5)
stl_fit  <- stl(aapl_ts, s.window = "periodic", robust = TRUE)
stl_df   <- data.frame(
  Date     = aapl$Date,
  Observed = as.numeric(stl_fit$time.series[, "remainder"] +
                        stl_fit$time.series[, "trend"] +
                        stl_fit$time.series[, "seasonal"]),
  Trend    = as.numeric(stl_fit$time.series[, "trend"]),
  Seasonal = as.numeric(stl_fit$time.series[, "seasonal"]),
  Residual = as.numeric(stl_fit$time.series[, "remainder"])
)

trend_gain <- last(stl_df$Trend) - first(stl_df$Trend)
seas_amp   <- max(stl_df$Seasonal) - min(stl_df$Seasonal)
noise_sd   <- sd(stl_df$Residual)

# One panel per component
make_decomp_panel <- function(df, y_col, title, subtitle, caption, color) {
  ggplot(df, aes_string(x = "Date", y = y_col)) +
    geom_line(color = color, linewidth = 0.85) +
    scale_x_date(date_labels = "%b '%y", date_breaks = "2 months") +
    labs(title = title, subtitle = subtitle, caption = caption,
         x = NULL, y = NULL) +
    theme_stock()
}

p_obs <- make_decomp_panel(stl_df, "Observed",
  "① Observed Price",
  "The raw closing price — trend + seasonality + noise all mixed together.",
  "This is what you see on any stock chart.", COL$aapl)

p_tr  <- make_decomp_panel(stl_df, "Trend",
  "② Underlying Trend",
  sprintf("Smoothed long-run direction. Total change: %+.2f USD over the year.", trend_gain),
  "With noise removed, the true direction of the stock becomes clear.", COL$ma50)

p_sea <- make_decomp_panel(stl_df, "Seasonal",
  "③ Weekly Seasonal Pattern",
  sprintf("Repeating weekly rhythm. Amplitude: %.2f USD (peak-to-trough).", seas_amp),
  "Stocks often have a slight Mon–Fri pattern (e.g. Monday dip, Friday rally).", COL$ma20)

p_res <- make_decomp_panel(stl_df, "Residual",
  "④ Residual Noise",
  sprintf("Random shocks not explained by trend or seasonality. Std Dev: %.2f USD.", noise_sd),
  "Spikes = unexpected news / earnings events. Cannot be predicted by trend models.", COL$neutral)

stl_plot <- (p_obs / p_tr / p_sea / p_res) +
  plot_annotation(
    title   = "AAPL Price — STL Decomposition (Trend + Seasonal + Noise)",
    caption = "STL = Seasonal and Trend decomposition using Loess. Splits one series into three interpretable parts.",
    theme   = theme(plot.title = element_text(face = "bold", size = 14))
  )

print(stl_plot)
ggsave("03_stl_decomposition.png", stl_plot, width = 13, height = 12, dpi = 140)
cat("  Chart saved → 03_stl_decomposition.png\n")

insight_box("Reading the STL decomposition", c(
  "Trend panel    : The true direction of the stock, cleaned of noise.",
  sprintf("  → AAPL's underlying trend moved %+.2f USD over the year.", trend_gain),
  sprintf("Seasonal panel : Weekly cycle of amplitude %.2f USD.", seas_amp),
  "  → Small effect — regular Mon–Fri trading pattern.",
  sprintf("Residual panel : Random noise, std dev = %.2f USD.", noise_sd),
  "  → Large spikes = news events, earnings, macro shocks.",
  "  → These cannot be predicted — they are pure randomness."
))

# ══════════════════════════════════════════════════════════════
#  STEP 6 — STATIONARITY TEST (ADF)
# ══════════════════════════════════════════════════════════════
section_header("6", "Stationarity Test (ADF)",
               "Does the price have a stable mean? Forecasting requires this.")

run_adf <- function(series, label) {
  result <- adf.test(series)
  status <- if (result$p.value < 0.05) "STATIONARY ✓" else "NON-STATIONARY ✗"
  cat(sprintf("  %-35s | ADF = %7.4f | p = %.4f | %s\n",
              label, result$statistic, result$p.value, status))
  result$p.value
}

cat("\n")
p_level <- run_adf(aapl$Close,                     "AAPL Close (raw price)")
p_diff  <- run_adf(diff(aapl$Close),               "AAPL Close (1st difference)")

insight_box("Why does stationarity matter for forecasting?", c(
  "ARIMA models need the series to have a stable, flat average.",
  "Raw stock prices are NOT stationary — they drift up or down.",
  "  → ADF p-value > 0.05 = non-stationary (confirmed above).",
  "First differencing (today minus yesterday) removes the drift.",
  "  → ADF p-value < 0.05 = stationary ✓ — now safe to model.",
  "This tells us AAPL is an I(1) process — integrated of order 1."
))

# ══════════════════════════════════════════════════════════════
#  STEP 7 — FORECASTING MODELS
# ══════════════════════════════════════════════════════════════
section_header("7", "Forecasting Models",
               "ARIMA, ETS and NNAR trained on 222 days, tested on last 30")

N_TEST  <- 30
n_train <- nrow(aapl) - N_TEST
train_ts <- ts(aapl$Close[1:n_train],          frequency = 5)
test_vals <- aapl$Close[(n_train + 1):nrow(aapl)]
test_dates <- aapl$Date[(n_train + 1):nrow(aapl)]

# ── ARIMA ────────────────────────────────────────────────────
cat("\n  Fitting ARIMA...\n")
arima_model <- auto.arima(train_ts, seasonal = TRUE,
                           stepwise = FALSE, approximation = FALSE)
arima_fc    <- forecast(arima_model, h = N_TEST)
arima_pred  <- as.numeric(arima_fc$mean)
arima_rmse  <- rmse(test_vals, arima_pred)
arima_mae   <- mae(test_vals,  arima_pred)
arima_mape  <- mape(test_vals, arima_pred) * 100
cat(sprintf("  ARIMA %s  →  RMSE: %.4f  MAE: %.4f  MAPE: %.2f%%\n",
            as.character(arima_model), arima_rmse, arima_mae, arima_mape))

# ── ETS ──────────────────────────────────────────────────────
cat("  Fitting ETS...\n")
ets_model  <- ets(train_ts)
ets_fc     <- forecast(ets_model, h = N_TEST)
ets_pred   <- as.numeric(ets_fc$mean)
ets_rmse   <- rmse(test_vals, ets_pred)
ets_mae    <- mae(test_vals,  ets_pred)
ets_mape   <- mape(test_vals, ets_pred) * 100
cat(sprintf("  ETS (%s)  →  RMSE: %.4f  MAE: %.4f  MAPE: %.2f%%\n",
            ets_model$method, ets_rmse, ets_mae, ets_mape))

# ── NNAR ─────────────────────────────────────────────────────
cat("  Fitting NNAR (neural network)...\n")
nnar_model <- nnetar(train_ts, repeats = 20)
nnar_fc    <- forecast(nnar_model, h = N_TEST)
nnar_pred  <- as.numeric(nnar_fc$mean)
nnar_rmse  <- rmse(test_vals, nnar_pred)
nnar_mae   <- mae(test_vals,  nnar_pred)
nnar_mape  <- mape(test_vals, nnar_pred) * 100
cat(sprintf("  NNAR  →  RMSE: %.4f  MAE: %.4f  MAPE: %.2f%%\n",
            nnar_rmse, nnar_mae, nnar_mape))

# ── Model comparison table ────────────────────────────────────
model_results <- tibble(
  Model = c("ARIMA", "ETS", "NNAR (Neural Net)"),
  RMSE  = round(c(arima_rmse, ets_rmse, nnar_rmse), 4),
  MAE   = round(c(arima_mae,  ets_mae,  nnar_mae),  4),
  MAPE  = round(c(arima_mape, ets_mape, nnar_mape), 2)
) %>% arrange(RMSE) %>%
  mutate(Rank = row_number(),
         Note = case_when(
           Rank == 1 ~ "⭐ Best model",
           Rank == 2 ~ "Runner-up",
           TRUE      ~ "Third place"
         ))

best_model <- model_results$Model[1]

section_header("7b", "Model Comparison Results")
print(model_results)

insight_box("How to read the model metrics", c(
  "RMSE (Root Mean Squared Error):",
  "  → Average prediction error in USD. Lower = better.",
  "  → Penalises large errors more heavily than small ones.",
  "MAE (Mean Absolute Error):",
  "  → Average absolute miss in USD. Easier to interpret.",
  "MAPE (Mean Absolute Percentage Error):",
  "  → Error as a % of the actual price. E.g. 1% = very good.",
  sprintf("Best overall model : %s (lowest RMSE)", best_model)
))

# ── Forecast vs Actual chart ──────────────────────────────────
hist_plot <- aapl %>% tail(60 + N_TEST) %>% head(60)
fc_df <- tibble(
  Date  = test_dates,
  Actual = test_vals,
  ARIMA  = arima_pred,
  ETS    = ets_pred,
  NNAR   = nnar_pred
) %>%
  pivot_longer(cols = c(Actual, ARIMA, ETS, NNAR),
               names_to = "Series", values_to = "Price")

fc_colors <- c(Actual = "#888780", ARIMA = "#E24B4A",
               ETS = "#BA7517", NNAR = "#378ADD")
fc_types  <- c(Actual = "solid",  ARIMA = "dashed",
               ETS = "dotdash", NNAR = "longdash")

p_fc_compare <- ggplot() +
  geom_line(data = hist_plot,
            aes(x = Date, y = Close, color = "Historical"),
            linewidth = 1.0) +
  geom_line(data = fc_df,
            aes(x = Date, y = Price, color = Series, linetype = Series),
            linewidth = 0.95) +
  geom_vline(xintercept = as.numeric(aapl$Date[n_train]),
             color = "#AAAAAA", linetype = "dotted", linewidth = 0.8) +
  annotate("text",
           x = aapl$Date[n_train] - 2,
           y = max(hist_plot$Close),
           label = "← Training data", hjust = 1, size = 2.8, color = "#666") +
  annotate("text",
           x = aapl$Date[n_train] + 2,
           y = max(hist_plot$Close),
           label = "Forecast zone →", hjust = 0, size = 2.8, color = "#666") +
  scale_color_manual(
    values = c(Historical = COL$aapl, fc_colors),
    name   = "Series"
  ) +
  scale_linetype_manual(
    values = c(Historical = "solid", fc_types),
    name   = "Series"
  ) +
  scale_y_continuous(labels = dollar_format()) +
  scale_x_date(date_labels = "%b %d", date_breaks = "2 weeks") +
  labs(
    title    = "AAPL — Three Models Predicting the Next 30 Days",
    subtitle = sprintf(
      "Gray = actual prices  |  Best model: %s (RMSE = %.4f USD)",
      best_model, model_results$RMSE[1]),
    caption  = paste0(
      "The closer a colored line stays to the gray 'Actual' line, the more accurate that model is.\n",
      "RMSE tells us: on average, predictions were off by that many dollars per day.\n",
      "A lower MAPE% = model predictions are a smaller % away from the real price."
    ),
    x = NULL, y = "Price (USD)"
  ) +
  theme_stock()

print(p_fc_compare)
ggsave("04_model_comparison.png", p_fc_compare, width = 13, height = 6, dpi = 140)
cat("  Chart saved → 04_model_comparison.png\n")

# ── Performance bar chart ─────────────────────────────────────
perf_long <- model_results %>%
  select(Model, RMSE, MAE, MAPE) %>%
  pivot_longer(cols = c(RMSE, MAE, MAPE), names_to = "Metric", values_to = "Value")

p_perf <- ggplot(perf_long, aes(x = reorder(Model, Value), y = Value,
                                 fill = Model == best_model)) +
  geom_bar(stat = "identity", width = 0.55, show.legend = FALSE) +
  geom_text(aes(label = round(Value, 2)), vjust = -0.4, size = 3, fontface = "bold") +
  scale_fill_manual(values = c(`TRUE` = "#1D9E75", `FALSE` = "#B4B2A9")) +
  facet_wrap(~Metric, scales = "free_y") +
  labs(
    title    = "Model Performance Comparison — Lower Is Better",
    subtitle = sprintf("Green bar = best model (%s). Gray = other models.", best_model),
    caption  = "RMSE & MAE in USD. MAPE in %. Smaller values = predictions closer to real prices.",
    x = NULL, y = "Error Value"
  ) +
  theme_stock() +
  theme(legend.position = "none",
        strip.text = element_text(face = "bold"))

print(p_perf)
ggsave("05_model_performance.png", p_perf, width = 11, height = 5, dpi = 140)
cat("  Chart saved → 05_model_performance.png\n")

# ══════════════════════════════════════════════════════════════
#  STEP 8 — 30-DAY FUTURE FORECAST
# ══════════════════════════════════════════════════════════════
section_header("8", "30-Day Future Price Forecast",
               "Using the full dataset to forecast the next 30 trading days")

full_ts    <- ts(aapl$Close, frequency = 5)
final_model <- auto.arima(full_ts, stepwise = FALSE, approximation = FALSE)
future_fc   <- forecast(final_model, h = 30, level = c(80, 95))

future_dates <- {
  d <- max(aapl$Date) + 1
  out <- c()
  while (length(out) < 30) {
    if (!weekdays(d) %in% c("Saturday", "Sunday")) out <- c(out, as.character(d))
    d <- d + 1
  }
  as.Date(out)
}

forecast_df <- tibble(
  Date      = future_dates,
  Forecast  = as.numeric(future_fc$mean),
  Lo_80     = as.numeric(future_fc$lower[, 1]),
  Hi_80     = as.numeric(future_fc$upper[, 1]),
  Lo_95     = as.numeric(future_fc$lower[, 2]),
  Hi_95     = as.numeric(future_fc$upper[, 2])
)

hist60 <- aapl %>% tail(60)

forecast_direction <- if (last(forecast_df$Forecast) > last(aapl$Close))
  "PREDICTED TO RISE ↑" else "PREDICTED TO FALL ↓"

price_now  <- last(aapl$Close)
price_fc30 <- last(forecast_df$Forecast)
price_chg  <- price_fc30 - price_now
pct_chg    <- price_chg / price_now * 100

p_future <- ggplot() +
  # Historical
  geom_line(data = hist60,
            aes(x = Date, y = Close), color = COL$aapl, linewidth = 1.1) +
  # 95% CI band
  geom_ribbon(data = forecast_df,
              aes(x = Date, ymin = Lo_95, ymax = Hi_95),
              fill = COL$forecast, alpha = 0.12) +
  # 80% CI band
  geom_ribbon(data = forecast_df,
              aes(x = Date, ymin = Lo_80, ymax = Hi_80),
              fill = COL$forecast, alpha = 0.22) +
  # Forecast line
  geom_line(data = forecast_df,
            aes(x = Date, y = Forecast),
            color = COL$forecast, linewidth = 1.3, linetype = "dashed") +
  # Start of forecast marker
  geom_vline(xintercept = as.numeric(max(aapl$Date)),
             color = "#AAAAAA", linetype = "dotted") +
  # Annotate current price
  annotate("point", x = max(aapl$Date), y = price_now,
           color = COL$aapl, size = 3) +
  annotate("label", x = max(aapl$Date) - 3, y = price_now,
           label = sprintf("Today\n$%.2f", price_now),
           size = 2.8, color = COL$aapl, fill = "#E6F1FB",
           label.size = 0.2, hjust = 1) +
  # Annotate end forecast price
  annotate("point", x = last(forecast_df$Date), y = price_fc30,
           color = COL$forecast, size = 3) +
  annotate("label", x = last(forecast_df$Date) + 2,
           y = price_fc30,
           label = sprintf("Day 30\n$%.2f\n(%+.1f%%)", price_fc30, pct_chg),
           size = 2.8, color = COL$forecast, fill = "#FBEAF0",
           label.size = 0.2, hjust = 0) +
  scale_y_continuous(labels = dollar_format()) +
  scale_x_date(date_labels = "%b %d", date_breaks = "2 weeks",
               expand = expansion(mult = c(0.02, 0.08))) +
  labs(
    title    = sprintf("AAPL — 30-Day Future Forecast  (%s)", forecast_direction),
    subtitle = sprintf(
      "Starting price: $%.2f  |  Forecast end: $%.2f  |  Expected change: %+.2f USD (%+.1f%%)",
      price_now, price_fc30, price_chg, pct_chg
    ),
    caption  = paste0(
      "Dark pink band = 80% confidence interval: the model is 80% sure the price stays in this range.\n",
      "Light pink band = 95% confidence interval: wider, more conservative estimate of the range.\n",
      "Dashed line = most likely price path (the model's best single guess).\n",
      "Note: No forecast is guaranteed. Wider bands = more uncertainty further into the future."
    ),
    x = NULL, y = "Price (USD)"
  ) +
  theme_stock()

print(p_future)
ggsave("06_future_forecast.png", p_future, width = 13, height = 6, dpi = 140)
cat("  Chart saved → 06_future_forecast.png\n")

# ══════════════════════════════════════════════════════════════
#  STEP 9 — FINAL PLAIN-ENGLISH SUMMARY REPORT
# ══════════════════════════════════════════════════════════════
section_header("9", "Final Summary — What Did We Find?",
               "Plain-English conclusions for the full analysis")

cat("\n")
cat("  ┌─────────────────────────────────────────────────────────────┐\n")
cat("  │              ANALYSIS SUMMARY REPORT — AAPL                │\n")
cat("  ├─────────────────────────────────────────────────────────────┤\n")
cat(sprintf("  │  Data period    : %s → %s         │\n",
            min(aapl$Date), max(aapl$Date)))
cat(sprintf("  │  Trading days   : %d                                      │\n", nrow(aapl)))
cat("  ├─────────────────────────────────────────────────────────────┤\n")
cat("  │  TREND FINDINGS                                             │\n")
cat(sprintf("  │  • Overall direction  : %-35s│\n", direction))
cat(sprintf("  │  • Daily price drift  : %+.4f USD per day                │\n", slope))
cat(sprintf("  │  • Year-end change    : %+.2f USD (linear estimate)      │\n", slope * 252))
cat(sprintf("  │  • Dominant trend     : %-35s│\n",
            names(which.max(table(aapl$Trend)))))
cat("  ├─────────────────────────────────────────────────────────────┤\n")
cat("  │  MOMENTUM SIGNALS                                           │\n")
last_rsi  <- last(na.omit(aapl$RSI))
last_macd <- last(na.omit(aapl$MACD_val))
last_sig  <- last(na.omit(aapl$MACD_sig))
rsi_status <- case_when(
  last_rsi > 70 ~ "OVERBOUGHT — potential pullback",
  last_rsi < 30 ~ "OVERSOLD   — potential bounce",
  TRUE          ~ "NEUTRAL    — no extreme signal"
)
macd_status <- if (last_macd > last_sig) "BULLISH — upward momentum" else "BEARISH — downward momentum"
cat(sprintf("  │  • RSI (last value)   : %.1f → %-30s│\n", last_rsi, rsi_status))
cat(sprintf("  │  • MACD signal        : %-36s│\n", macd_status))
cat(sprintf("  │  • Golden crosses     : %-2d events (bullish momentum shifts)   │\n", nrow(crossover_up)))
cat(sprintf("  │  • Death crosses      : %-2d events (bearish momentum shifts)   │\n", nrow(crossover_dn)))
cat("  ├─────────────────────────────────────────────────────────────┤\n")
cat("  │  FORECAST                                                   │\n")
cat(sprintf("  │  • Best model         : %-35s│\n", best_model))
cat(sprintf("  │  • Model error (RMSE) : $%-34.4f│\n", model_results$RMSE[1]))
cat(sprintf("  │  • Model error (MAPE) : %-34s│\n",
            sprintf("%.2f%% avg miss as %% of price", model_results$MAPE[1])))
cat(sprintf("  │  • 30-day outlook     : %-35s│\n", forecast_direction))
cat(sprintf("  │  • Price in 30 days   : $%.2f  (from $%.2f today)       │\n",
            price_fc30, price_now))
cat(sprintf("  │  • Expected change    : %+.2f USD  (%+.1f%%)                 │\n",
            price_chg, pct_chg))
cat("  └─────────────────────────────────────────────────────────────┘\n")

cat("\n  Output files generated:\n")
cat("    01_eda_dashboard.png      — Overview of all 4 stocks\n")
cat("    02_trend_analysis.png     — AAPL trend: MAs, RSI, MACD\n")
cat("    03_stl_decomposition.png  — Trend vs seasonal vs noise\n")
cat("    04_model_comparison.png   — ARIMA vs ETS vs NNAR forecast\n")
cat("    05_model_performance.png  — RMSE / MAE / MAPE bar chart\n")
cat("    06_future_forecast.png    — 30-day future price forecast\n")

cat("\n  ✅ Analysis complete.\n\n")
```
